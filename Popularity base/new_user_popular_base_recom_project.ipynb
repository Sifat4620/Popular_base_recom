```python
import pandas as pd
import json

#  path to your JSON file
json_file_path = '../rokomari_all_books.json'
file_encoding = 'utf-8'  

# Open the JSON file for read with the encoding
with open(json_file_path, 'r', encoding=file_encoding) as json_file:

    data = json.load(json_file)
```

```python
data
```

```python
### Title
### Author
### Category
### Original Price
### Sale Price
### Stock Availability
### No of copies available
### No of Reviews

title=[]
author=[]
category =[]
original_price =[]
sale_price = []
stock = []
no_copies_available = []
no_reviews=[]

for i in data: 
      title.append(i['Title'])
      author.append(i['Author'])
      category.append(i['Category'])
      original_price.append(i['Original Price'])
      sale_price.append(i['Sale Price'])
      stock.append(i['Stock Availability'])
      no_copies_available.append(i['No of copies available'])
      no_reviews.append(i['No of Reviews'])
        
    
df={'title':title,'author':author,'category':category,'original_price':original_price,'sale_price':sale_price,'stock':stock,
    'no_copies_available':no_copies_available,'no_reviews':no_reviews}  

df=pd.DataFrame(df)
```

```python
df['no_reviews'] = pd.to_numeric(df['no_reviews'], errors='coerce')
```
